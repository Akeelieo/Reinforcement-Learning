```python
import random
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import gymnasium as gym
from torch.distributions import Categorical

# Import from supplied files
from dqn_starter import train as train_dqn, run_multiple_seeds as dqn_seeds, QNetwork, ReplayBuffer, DQNAgent
from policy_gradient_starter import train_reinforce, run_multiple_seeds as pg_seeds
```

```python
# Cell 2 - DQN over 5 seeds
print("Running DQN over 5 seeds...")
dqn_returns = dqn_seeds(
    env_name="CartPole-v1",
    n_seeds=5,
    n_episodes=1000,
    use_target_network=True,
)
print(f"DQN done! Final 100-episode mean: {dqn_returns.mean(axis=0)[-100:].mean():.1f}")
```

```python
# Cell 3 - REINFORCE over 5 seeds
print("Running REINFORCE over 5 seeds...")
reinforce_returns = pg_seeds(
    train_reinforce,
    n_seeds=5,
    env_name="CartPole-v1",
    n_episodes=1000,
    use_baseline=True,
)
print(f"REINFORCE done! Final 100-episode mean: {reinforce_returns.mean(axis=0)[-100:].mean():.1f}")
```

```python
# Cell 4 - Main comparison plot
window = 20
kernel = np.ones(window) / window

def smooth(arr):
    return np.convolve(arr, kernel, mode='valid')

dqn_mean_s = smooth(dqn_returns.mean(axis=0))
dqn_std_s = smooth(dqn_returns.std(axis=0))
reinforce_mean_s = smooth(reinforce_returns.mean(axis=0))
reinforce_std_s = smooth(reinforce_returns.std(axis=0))
x = np.arange(len(dqn_mean_s))

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, dqn_mean_s, color='steelblue', linewidth=1.5,
        label='DQN (with target network)')
ax.fill_between(x, dqn_mean_s - dqn_std_s,
                dqn_mean_s + dqn_std_s,
                alpha=0.2, color='steelblue')

ax.plot(x, reinforce_mean_s, color='tomato', linewidth=1.5,
        label='REINFORCE (with baseline)')
ax.fill_between(x, reinforce_mean_s - reinforce_std_s,
                reinforce_mean_s + reinforce_std_s,
                alpha=0.2, color='tomato')

ax.axhline(475, color='darkgreen', linestyle='--', linewidth=1.5,
           label='Solved threshold (475)')

ax.set_xlabel('Episode')
ax.set_ylabel('Return')
ax.set_title('DQN vs REINFORCE — CartPole-v1\n'
             '5 seeds, ±1 std shading, 20-episode moving average')
ax.legend(fontsize=9, loc='upper left')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.savefig("dqn_vs_reinforce.png", dpi=140, bbox_inches="tight")
plt.show()

print(f"DQN final 100-episode mean:       {dqn_returns.mean(axis=0)[-100:].mean():.1f}")
print(f"REINFORCE final 100-episode mean: {reinforce_returns.mean(axis=0)[-100:].mean():.1f}")
```

```python
# DQN without target network
print("Running DQN without target network over 5 seeds...")
dqn_no_target = dqn_seeds(
    env_name="CartPole-v1",
    n_seeds=5,
    n_episodes=1000,
    use_target_network=False,
)
print(f"Done! Final 100-episode mean: {dqn_no_target.mean(axis=0)[-100:].mean():.1f}")
```

```python
# Ablation plot
dqn_nt_mean_s = smooth(dqn_no_target.mean(axis=0))
dqn_nt_std_s = smooth(dqn_no_target.std(axis=0))

fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(x, dqn_mean_s, color='steelblue', linewidth=1.5,
        label='DQN (with target network)')
ax.fill_between(x, dqn_mean_s - dqn_std_s,
                dqn_mean_s + dqn_std_s,
                alpha=0.2, color='steelblue')

ax.plot(x, dqn_nt_mean_s, color='orange', linewidth=1.5,
        label='DQN (no target network)')
ax.fill_between(x, dqn_nt_mean_s - dqn_nt_std_s,
                dqn_nt_mean_s + dqn_nt_std_s,
                alpha=0.2, color='orange')

ax.axhline(475, color='darkgreen', linestyle='--', linewidth=1.5,
           label='Solved threshold (475)')

ax.set_xlabel('Episode')
ax.set_ylabel('Return')
ax.set_title('DQN Ablation: Target Network vs No Target Network\n'
             'CartPole-v1, 5 seeds, ±1 std shading')
ax.legend(fontsize=9, loc='upper left')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.savefig("dqn_ablation.png", dpi=140, bbox_inches="tight")
plt.show()

print(f"DQN with target final mean:    {dqn_returns.mean(axis=0)[-100:].mean():.1f}")
print(f"DQN no target final mean:      {dqn_no_target.mean(axis=0)[-100:].mean():.1f}")
```
