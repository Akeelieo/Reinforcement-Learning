```python
# Make sure gridworld.py is in the same folder as this notebook
import numpy as np
import matplotlib.pyplot as plt
from gridworld import (
    GridWorld,
    compare_policies,
    evaluate_policy,
    plot_learning_curves,
    plot_policy,
    plot_value_function,
)

# Copy these functions from tabular_rl_starter.py
from tabular_rl_starter import (
    value_iteration,
    q_learning,
    sarsa,
    run_hyperparameter_study,
    plot_hyperparameter_study,
    summarise_policy_performance,
    greedy_action,
    epsilon_greedy,
    extract_policy_from_q,
    extract_policy_from_values,
)
```

```python
# Create environment
env = GridWorld(seed=0)
print(env.describe_layout())

# Run value iteration
V_opt, pi_opt, n_iters = value_iteration(env)
print(f"Converged in {n_iters} iterations")

# Plot value function and optimal policy
fig, axes = plt.subplots(1, 2, figsize=(11, 5))
plot_value_function(env, V_opt, title="Optimal Value Function", ax=axes[0])
plot_policy(env, pi_opt, title="Optimal Policy", ax=axes[1])
fig.tight_layout()
plt.savefig("value_iteration.png", dpi=140, bbox_inches="tight")
plt.show()
```

```python
env_q = GridWorld(seed=0)
Q_q, pi_q, returns_q = q_learning(
    env_q, 
    n_episodes=5000, 
    alpha=0.1, 
    epsilon=0.1, 
    seed=0
)

# Evaluate policies
returns_opt = evaluate_policy(env, pi_opt, n_episodes=1000, seed=123)
returns_q_eval = evaluate_policy(env_q, pi_q, n_episodes=1000, seed=123)

opt_mean = returns_opt.mean()
q_mean = returns_q_eval.mean()

# Smooth with moving average
window = 100
kernel = np.ones(window) / window
smoothed = np.convolve(returns_q, kernel, mode='valid')
x = np.arange(len(smoothed))

fig, ax = plt.subplots(figsize=(10, 5))

# Raw returns clipped
ax.plot(np.clip(returns_q, -0.05, None), 
        color='lightblue', alpha=0.4, linewidth=0.5, 
        label='Raw episode return (clipped at -0.05)')

# Smoothed moving average
ax.plot(x, smoothed, color='steelblue', linewidth=1.5, 
        label='100-episode moving average')

# Optimal policy line
ax.axhline(opt_mean, color='darkgreen', linestyle='--', linewidth=1.5,
           label=f'Optimal policy mean return ({opt_mean:.4f})')

# Q-learning greedy mean line
ax.axhline(q_mean, color='purple', linestyle='-.', linewidth=1.5,
           label=f'Q-learning greedy mean return ({q_mean:.4f})')

# Annotation
ax.annotate('Fast initial\nlearning (~500 ep)', 
            xy=(500, smoothed[500-window//2]),
            xytext=(800, 0.57),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=9)

ax.set_xlabel('Episode')
ax.set_ylabel('Discounted Episode Return')
ax.set_ylim(0.5, 0.7)
ax.set_title(f'(b) Q-Learning Learning Curve — GridWorld 5×5\n'
             f'ε = 0.1, α = 0.1, γ = 0.95, seed = 0')
ax.legend(fontsize=9, loc='lower right')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.savefig("q_learning_curve.png", dpi=140, bbox_inches="tight")
plt.show()

print(f"Optimal policy mean return: {opt_mean:.4f}")
print(f"Q-learning greedy mean return: {q_mean:.4f}")
agreement = np.sum(pi_q == pi_opt)
print(f"Policy agreement: {agreement}/23 states ({100*agreement/23:.1f}%)")
```

```python
# Run SARSA with same hyperparameters
env_s = GridWorld(seed=0)
Q_s, pi_s, returns_s = sarsa(
    env_s, 
    n_episodes=5000, 
    alpha=0.1, 
    epsilon=0.1, 
    seed=0
)

# Evaluate all policies
returns_opt = evaluate_policy(env, pi_opt, n_episodes=1000, seed=123)
returns_q_eval = evaluate_policy(env_q, pi_q, n_episodes=1000, seed=123)
returns_s_eval = evaluate_policy(env_s, pi_s, n_episodes=1000, seed=123)

opt_mean = returns_opt.mean()
q_mean = returns_q_eval.mean()
s_mean = returns_s_eval.mean()

# Policy agreement
compare_policies(env_q, pi_q, pi_opt, "Q-learning", "Optimal")
compare_policies(env_s, pi_s, pi_opt, "SARSA", "Optimal")

# Smooth both curves
window = 100
kernel = np.ones(window) / window
smoothed_q = np.convolve(returns_q, kernel, mode='valid')
smoothed_s = np.convolve(returns_s, kernel, mode='valid')
x = np.arange(len(smoothed_q))

fig, ax = plt.subplots(figsize=(8, 4))
# Raw returns
ax.plot(np.clip(returns_q, -0.05, None), 
        color='lightblue', alpha=0.3, linewidth=0.5)
ax.plot(np.clip(returns_s, -0.05, None), 
        color='lightsalmon', alpha=0.3, linewidth=0.5)

# Smoothed curves
ax.plot(x, smoothed_q, color='steelblue', linewidth=1.5, 
        label='Q-learning')
ax.plot(x, smoothed_s, color='tomato', linewidth=1.5, 
        label='SARSA')

# Reference lines
ax.axhline(opt_mean, color='darkgreen', linestyle='--', linewidth=1.5,
           label=f'Optimal ({opt_mean:.4f})')
ax.axhline(q_mean, color='steelblue', linestyle='-.', linewidth=1,
           label=f'Q-learning greedy ({q_mean:.4f})')
ax.axhline(s_mean, color='tomato', linestyle='-.', linewidth=1,
           label=f'SARSA greedy ({s_mean:.4f})')

ax.set_ylim(0.35, 0.75)
ax.set_xlabel('Episode')
ax.set_ylabel('Discounted Episode Return')
ax.set_title('(c) Q-Learning vs SARSA — GridWorld 5×5\n'
             'ε = 0.1, α = 0.1, γ = 0.95, seed = 0')
ax.legend(fontsize=8, loc='lower right')
ax.grid(True, alpha=0.3)

fig.tight_layout()
plt.savefig("q_vs_sarsa.png", dpi=140, bbox_inches="tight")
plt.show()

print(f"Optimal policy mean return:     {opt_mean:.4f} ± {returns_opt.std():.4f}")
print(f"Q-learning greedy mean return:  {q_mean:.4f} ± {returns_q_eval.std():.4f}")
print(f"SARSA greedy mean return:       {s_mean:.4f} ± {returns_s_eval.std():.4f}")
```

```python
# Three policy comparison plot
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
plot_policy(env, pi_opt, title='Optimal Policy $\pi^*$', ax=axes[0])
plot_policy(env_q, pi_q, title='Q-learning Policy', ax=axes[1])
plot_policy(env_s, pi_s, title='SARSA Policy', ax=axes[2])
fig.tight_layout()
plt.savefig("policy_comparison.png", dpi=140, bbox_inches="tight")
plt.show()

# Quantitative comparison
print(f"{'Policy':<20} {'Mean Return':>12} {'Std':>10} {'Agreement':>12}")
print("-" * 56)
print(f"{'Optimal':<20} {opt_mean:>12.4f} {returns_opt.std():>10.4f} {'---':>12}")
print(f"{'Q-learning':<20} {q_mean:>12.4f} {returns_q_eval.std():>10.4f} {'19/23 (82.6%)':>12}")
print(f"{'SARSA':<20} {s_mean:>12.4f} {returns_s_eval.std():>10.4f} {'16/23 (69.6%)':>12}")
```
